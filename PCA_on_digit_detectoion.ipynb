```
import numpy as np
import matplotlib.pyplot as plt
```

```
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2,2),rng.randn(2,200)).T
plt.scatter(X[:,0],X[:,1])
plt.axis('equal')
```

```
from sklearn.decomposition import PCA
pca = PCA(n_components =2)
pca.fit(X)
```

```
print("the PCA components are :",pca.components_)
print("the PCA variance is ",pca.explained_variance_)
```

```
def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops=dict(arrowstyle='->',
                    linewidth=2,
                    shrinkA=0, shrinkB=0)
    ax.annotate('', v1, v0, arrowprops=arrowprops)

# plot data
plt.scatter(X[:, 0], X[:, 1], alpha=0.6)
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)
plt.axis('equal');
```

```
pca =  PCA(n_components=1)
pca.fit(X)
X_pca = pca.transform(X)
print("Original shape",X.shape)
print("Transformed shape",X_pca.shape)
```

```
X_new = pca.inverse_transform(X_pca)
plt.scatter(X[:,0],X[:,1],alpha = 0.3)
plt.scatter(X_new[:,0],X_new[:,1],alpha=0.7)
plt.axis("equal")
print(X_new.shape)
```

#PCA on digit recognition

in this section i am going to import the data and visualize it

```
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
#axs = ['ax' + str(i) for i in range(1,401)]
axs = [i for i in range(1,401)]
fig = plt.figure(figsize=(15,10))
i=1
for ax in axs:
  ax = fig.add_subplot(20,20,i)
  ax.imshow(digits.data[i].reshape(8,8))
  ax.axis("off")
  i = i+1
  
plt.imshow(digits.data[3].reshape(8,8))
```

#Applying the PCA method on the data
and now its is the PCA 's turn to be applied on the data'

```
pca = PCA(5)
pca_digit = pca.fit_transform(digits.data)
print(digits.data.shape)
print(pca_digit.shape)
```

Visualizing the 'pca' ed data


```
new_digits = pca.inverse_transform(pca_digit)
print(new_digits.shape)
axess = ['axx' + str(i) for i in range(1,401-20)]
i=0
print("and you can compare the 'pca' ed data with the original data   ")
fig = plt.figure(figsize=(15,10))
for axx in axess:
  axx = fig.add_subplot(20,19,i+1)
  if(i%2):
     axx.imshow(new_digits[i].reshape(8,8))
  else:
     axx.imshow(digits.data[i-1].reshape(8,8))
  axx.axis("off")
  i = i+1


```
